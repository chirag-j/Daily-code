```python
import numpy as np
```

```python
cb = np.zeros((8,8))
cb[0,1] = 2
```

```python
def validPos(cr):
    x, y = cr
    ap = np.array([[x+2,y-1], [x+2,y+1], [x-1,y+2], [x+1,y+2], [x-2,y-1], [x-2,y+1], [x-1,y-2], [x+1,y-2]])
    temp = np.logical_and(ap>=0,ap<8)
    vp = ap[np.logical_and(temp[:,0],temp[:,1])]    # discard all rows with one or more false
    return vp
validPos([0,1])
```

```python
def tourDeKnight(cr):
    if len(cb[cb==0]) == 0:
        return True
    vp = validPos(cr)
    for loc in vp:
        if cb[loc[0],loc[1]] == 0:
            cb[loc[0],loc[1]] = 2
            cb[cr[0],cr[1]] = 1
#             print(cb)
            if tourDeKnight([loc[0],loc[1]]):
                return True
            else:
#                 print("backtracking...")
                cb[loc[0],loc[1]] = 0
                cb[cr[0],cr[1]] = 2
    return False
```

```python
tourDeKnight([0,1])
```
