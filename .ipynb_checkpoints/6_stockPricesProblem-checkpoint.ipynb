```python
import numpy as np
from pandas import DataFrame
```

```python
sp = np.array([[9,11,8,5,7,10,16,12,15,9,23,45,67,49,33,24,19]])
```

```python
pl = np.tril(sp.T - sp)
np.max(pl)
```

```python
rep = DataFrame(pl, columns = sp.flatten(), index =sp.flatten())
rep.columns.name = 'Enter at'
rep.index.name = 'Leave at'
```
